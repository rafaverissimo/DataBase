```python
%%sql
SELECT *
FROM sqlite_master
WHERE type='table';
```

```python
%%sql
SELECT *
  FROM facts
    LIMIT 5;
```

```python
%%sql 
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
    FROM facts;
```

```python
%%sql 
SELECT name
    FROM facts
    WHERE population=0
```

```python
%%sql 
SELECT name
    FROM facts
    WHERE population= (SELECT MAX(population) FROM facts)
```

```python
%%sql 
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
    FROM facts
    WHERE name NOT IN ("World","Antarctica")
```

```python
%%sql 
SELECT AVG(population) AS avg_population, AVG(area) AS avg_area
    FROM facts
    WHERE name NOT IN ("World","Antarctica")
```

```python
%%sql 
SELECT name, CAST(population AS FLOAT)/area AS Population_density
    FROM facts
    WHERE population>(SELECT AVG(population) AS avg_population FROM facts WHERE name NOT IN ("World","Antarctica"))
    AND area< (SELECT  AVG(area) AS avg_area FROM facts WHERE name NOT IN ("World","Antarctica"))
    ORDER BY Population_density DESC;
```

```python
%%sql
SELECT name, area_water
    FROM facts
    ORDER BY area_water DESC;
```

```python
%%sql
SELECT name, CAST(area_water AS FLOAT)/area AS WaterShare
    FROM facts
    ORDER BY WaterShare DESC;
```
